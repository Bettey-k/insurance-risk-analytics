```python
# 1. Setup and Imports
import sys
from pathlib import Path

# Add the project root to Python path
project_root = Path.cwd().parent  # Goes up one level from notebooks/
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

# Now import your modules
from src.data import load_data, preprocess_data
from src.visualization import plot_distributions, plot_correlations

# Standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

# Set up visualization
# Set up visualization
sns.set_theme(style="whitegrid", palette="husl")
%matplotlib inline


# Create output directories
output_dirs = ['visualizations', 'data/processed', 'reports']
for dir_path in output_dirs:
    os.makedirs(dir_path, exist_ok=True)

# 2. Load and Prepare Data
print(f"\n{datetime.now().strftime('%Y-%m-%d %H:%M:%S')} - Loading and preprocessing data...")
try:
    df_raw = load_data('../data/MachineLearningRating_v3.txt')  # Adjusted path
    print(f"Raw data loaded. Shape: {df_raw.shape}")
    
    df_processed = preprocess_data(df_raw)
    print(f"Data preprocessed. Shape: {df_processed.shape}")
    
    # Save processed data
    processed_data_path = 'data/processed/insurance_data_processed.parquet'
    df_processed.to_parquet(processed_data_path, index=False)
    print(f"Processed data saved to {processed_data_path}")
    
except Exception as e:
    print(f"Error loading data: {e}")
    raise

# 3. Basic Data Exploration
print("\n--- Basic Data Exploration ---")
print("Dataset Shape:", df_processed.shape)
print("\nData Types:\n", df_processed.dtypes)
print("\nMissing Values:\n", df_processed.isnull().sum())

# 4. Distribution Analysis
print("\n--- Distribution Analysis ---")
dist_plot = plot_distributions(df_processed)
dist_plot.savefig('visualizations/distributions.png', bbox_inches='tight', dpi=300)
plt.close()

# 5. Correlation Analysis
print("\n--- Correlation Analysis ---")
corr_plot = plot_correlations(df_processed)
corr_plot.savefig('visualizations/correlation_matrix.png', bbox_inches='tight', dpi=300)
plt.close()

# 6. Categorical Analysis
print("\n--- Categorical Analysis ---")
categorical_cols = ['VehicleType', 'Gender', 'MaritalStatus']

for col in categorical_cols:
    if col in df_processed.columns:
        # Box plots for Loss Ratio by category
        plt.figure(figsize=(12, 6))
        sns.boxplot(data=df_processed, x=col, y='LossRatio')
        plt.title(f'Loss Ratio by {col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(f'visualizations/loss_ratio_by_{col.lower()}.png', bbox_inches='tight', dpi=300)
        plt.close()

print("\n=== EDA Complete ===")
print(f"Visualizations saved to: {os.path.abspath('visualizations/')}")
print(f"Processed data saved to: {os.path.abspath(processed_data_path)}")
```
